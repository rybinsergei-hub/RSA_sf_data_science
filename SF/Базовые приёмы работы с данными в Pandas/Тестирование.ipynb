```python
# Задание 6.8

# В наших данных присутствуют столбцы, которые дублируют информацию друг о друге: это столбцы с идентификатором и названием стартовой и конечной стоянки. 
# Удалите признаки идентификаторов стоянок. Сколько столбцов осталось?

citibike_mod = citibike_df.drop(['start station name','end station name'],axis=1,inplace=False)


```

```python
# Задание 6.9

# Замените признак birth year на более понятный признак возраста клиента age. 
# Годом отсчёта возраста выберите 2018 год. Столбец birth year удалите из таблицы. 
# Сколько поездок совершено клиентами старше 60 лет?

citibike_mod['age']= 2018 - citibike_mod['birth year']
citibike_mod2 = citibike_mod.drop('birth year', axis=1)
old_bikers = citibike_mod2[citibike_mod['age'] > 60].shape[0]

print(citibike_mod2['age'].value_counts())
print(old_bikers)





```
